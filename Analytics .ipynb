```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/data/PreprocessedDataCoSupplyChainDataset.csv", encoding='latin1')
data.head()
```

## According to the heatmap from Correlation, we can see that product price has a high correlation with sales, order item total and sales per customer

### Checking which region has the most sales

```python
market_group = data.groupby('Market')

plt.figure(1)
market_group['Sales per customer'].sum().sort_values(ascending=False).plot.bar(figsize=(12,6), color=['black'], title="Total sales against all markets")
```

From the above histogram we can observe that European market has the highest number of sales whereas Africa has the lowest number of sales.

```python
region_group = data.groupby('Order Region')

plt.figure(2)
region_group['Sales per customer'].sum().sort_values(ascending=False).plot.bar(figsize=(12,6), color=['black'], title="Total sales against all regions")
```

From the above histogram, we can see that Western Europe region and Central America region recorded the highest sales whereas Central Asia recorded the lowest sales.

### Check which product category has the highest sales

```python
category = data.groupby('Category Name')
plt.figure(1)

category['Sales per customer'].sum().sort_values(ascending=False).plot.bar(figsize=(12,6), color=['black'], title="Total sales against all categories")
```

From the above histogram we can see that Fishing has the highest number of sales followed by the Cleats.

```python
plt.figure(2)

category['Sales per customer'].mean().sort_values(ascending=False).plot.bar(figsize=(12,6), color=['black'],title="Average sales against all categories")
```

```python
plt.figure(3)

category['Product Price'].mean().sort_values(ascending=False).plot.bar(figsize=(12,6), color=['black'], title="Average price against all categories")
```

From the histogram figure 2,3, we can observe that products with the highest prices on average are the ones that has the most sold products on average. 
Computers have close to 1400 sales despite the average price of a computer being close to $1500.

### Checking how pricing impacting the sales for all the products

```python
data.plot(x='Product Price', y='Sales per customer',linestyle='dotted',
     markerfacecolor='blue', markersize=12, color=['black']) 
plt.title('Product Price against Sales per customer')
plt.xlabel('Product Price')
plt.ylabel('Sales per customer') 
plt.show()
```

As you can see from the above graph, that price has a linear relation with sales.
